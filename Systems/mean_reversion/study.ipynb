```python
from mean_reversion import evaluate_against_symbol
import pandas as pd
from pathlib import Path
import os

entries = []

parent = Path(os.path.abspath('')).parents[1]
stocks_df = pd.read_csv(os.path.join(parent, 'Stock Info.csv'))
for i, symbol in enumerate(stocks_df['symbol']):
    history_df = pd.read_csv(os.path.join(parent, 'Stock History', f'{symbol}.csv'), index_col='Date', parse_dates=True)
    entries.append(evaluate_against_symbol(history_df))
    if i % 100 == 0:
        series = pd.Series(entries).sort_values(ascending=False)[:min(int(i/10), 100)]
        mean = series.mean()
        stddev = series.std()
        print(r'Top 10% stats by stock', i)
        print(f'Mean: {mean:.2f}')
        print(f'Stddev: {stddev:.2f}')
        print(f'Mean-2stddev: {mean - 2 * stddev:.2f}')
```
